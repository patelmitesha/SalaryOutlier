```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_excel('TBPY_SALARY.xlsx')
num_array = data.to_numpy()
salary_code_amount = num_array[:,[1,2,3]]
print(salary_code_amount)
salary_code_amount.size
```

```
fig = plt.figure(figsize =(20, 100)) 
  
# Creating plot 
sns.boxplot(y=salary_code_amount[:,1],x=salary_code_amount[:,2]) 

# Add jitter with the swarmplot function.
#sns.swarmplot(y=salary_code_amount[:,1],x=salary_code_amount[:,2], color="black")
  
# show plot 
plt.show() 
```

```
# filter by salitem code 1
salary_filterbysalitemcode = salary_code_amount[(salary_code_amount[:,1] != 428)]
salary_filterbysalitemcode = salary_filterbysalitemcode[(salary_filterbysalitemcode[:,1] != 419)]
print(salary_filterbysalitemcode)
```

```
fig = plt.figure(figsize =(20, 100)) 
  
# Creating plot 
sns.boxplot(y=salary_filterbysalitemcode[:,1],x=salary_filterbysalitemcode[:,2]) 

# Add jitter with the swarmplot function.
#sns.swarmplot(y=salary_code_amount[:,1],x=salary_code_amount[:,2], color="black")
  
# show plot 
plt.show() 
```
